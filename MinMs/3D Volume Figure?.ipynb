## Figuring out how to plot the MinMs targets in 3D

```python
#Importing some modules to help out with the plotting 
from mpl_toolkits import mplot3d #This contains a lot of the 3D stuff that we can play around with 
%matplotlib notebook

#Standard imports below 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fig = plt.figure() #Creating the figure 
ax = plt.axes(projection='3d') #This plots the axes that we see below. The projection = 3d is important as it 
#tells python to plot the axes in 3D space
```

```python
#Importing the data from the excel sheets and into arrays 
new_data = pd.read_csv('MinMs_Sample_Properties - Gaia Properties.csv', delimiter = '\t')
old_data = pd.read_csv('MinMs_Sample_Properties - MinMs_old.csv', delimiter = '\t')
```

```python
new_data.head()
```

```python
ax = plt.axes(projection='3d')

#plot?

lon = new_data['_RAJ2000']
lat = new_data['_DEJ2000']
radius = new_data['Distance']

ax.plot(lon, lat, radius, '.')
ax.set_xlabel('RA (ICRS)')
ax.set_ylabel('Dec (ICRS)')
ax.set_zlabel('Radius?')
```

```python
ax = plt.axes(projection='3d')

#plot?

lon = old_data['RAJ2000  deg']
lat = old_data['DEJ2000  deg']
radius = old_data['Dist  pc']

ax.plot(lon, lat, radius, '.')
ax.set_xlabel('RA (ICRS)')
ax.set_ylabel('Dec (ICRS)')
ax.set_zlabel('Radius?')
plt.savefig('Old Data.png')
```

```python
#Trying to see if its possible to make a boundary at 15pc...
num_pts = 100
indices = np.arange(0, num_pts, dtype=float)

phi = np.arccos(1 - 2*indices/num_pts)
theta = np.pi * (1 + 5**0.5) * indices

x, y, z = np.cos(theta) * np.sin(phi), np.sin(theta) * np.sin(phi), np.cos(phi);

plt.figure().add_subplot(111,projection='3d').scatter(x, y, z);
plt.show()
```
