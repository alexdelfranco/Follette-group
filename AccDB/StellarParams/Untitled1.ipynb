```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import seaborn as sns; sns.set(style="white", color_codes=True)
import pandas as pd
import scipy.stats as st
```

```python
BD = pd.read_csv("Accretion Rate Database - Brown Dwarfs.csv",encoding='windows-1254')
```

```python
BD=BD.dropna(subset=['Re-mass'])
```

```python
print(len(BD))
```

```python
plt.hist(BD["Re-mass"])
```

```python
remass = BD['Re-mass']
mass = BD['Object Mass M_Solar']
```

```python
a = remass - mass
plt.hist(remass)
```

```python
plt.hist(a,bins=30)
plt.title("Distribution of Mass Differences (Re-estimated - Original)")
plt.xlabel("Mass Difference ($M_\odot$)")
plt.savefig('dist',dpi=200)
```
